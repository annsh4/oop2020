На підприємстві функціонує система електронного документообігу. Система обробляє три типи документів: листи, накази, розпорядження. Кожен документ має: номер, дату, короткий зміст. Крім того, в листах вказується тип: вхідний/вихідний і кореспондент. А в наказах: підрозділ, строк виконання, і виконавець. А в розпорядженнях: працівник, період і робоче місце.

Вивести: 1. Перелік документів в системі електронного документообігу.                                           
         2. Зміст обраного документу за його номером. 

```python
class Documents:
    def __init__(self, number, date, summary):
        self.number = number
        self.date = date
        self.summary = summary
        
class Letters(Documents):
    def __init__(self, number, date, summary, letter_type, correspon):
        Documents.__init__(self, number, date, summary)
        self.letter_type = letter_type
        self.correspon = correspon
        
class Orders(Documents):
    def __init__(self, number, date, summary, subdiv, deadline, executant):
        Documents.__init__(self, number, date, summary)
        self.subdiv = subdiv
        self.deadline = deadline
        self.executant = executant
        
class Commands(Documents):
    def __init__(self, number, date, summary, worker, period, workplace):
        Documents.__init__(self, number, date, summary)
        self.worker = worker
        self.period = period
        self.workplace = workplace
        
class DocumentFlow:
    documents = dict()
    
    def add_document(self, document):
        self.documents[document.number] = document
        
    def get_values(self):
        return self.documents.values()

    def get_by_number(self, number):
        return self.documents[number].summary
    
documentFlow = DocumentFlow()
letter = Letters(1, "05.03.2021", "Лист за сплату електроенергії", "Вхідний", "Комунальні послуги")
order = Orders(2, "06.03.2021", "Наказ по складанню статистики", "Бухгалтерія", "06.04.2021", "Пітер Джонс")
command = Commands(3, "07.03.2021", "Розпорядження зібрати всіх на обговорення", "Наталі Портман", 
                   "10.03.2021", "Конференцзал")

documentFlow.add_document(letter)
documentFlow.add_document(order)
documentFlow.add_document(command)

print("Перелік документів в системі електронного документообігу:")
for document in documentFlow.get_values():
    a = dict()
    a = document.__dict__
    for key, value in a.items():
        print(key, "-", value)
    print()
    
print("Зміст обраного документу за його номером:")    
print(documentFlow.get_by_number(2))
```

В інформаційній системі Інтернет-магазину зберігається інформація про номенклатуру товарів трьох типів: материнські плати, процесори, відеокарти. Кожен товар має: номенклатурний номер, найменування і вартість. Крім того, для материнських плат вказується: тип сокета, кількість процесорів, тип оперативної пам’яті, частота системної шини; для процесорів – тип сокета, кількість ядер, тактова частота, техпроцес; для відеокарт – тактова частота відеочипу, швидкість заповнення (филлрейт), кількість блоків текстурірування.

Вивести: 1. Повний перелік комплектуючих інформаційної системи Інтернет-магазину. 
         2. Детальну інформацію по товару за його номенклатурним номером. 

```python
class InternetStore:
    def __init__(self, nomencla_num, name, price):
        self.nomencla_num = nomencla_num
        self.name = name
        self.price = price
        
class Motherboards(InternetStore):
    def __init__(self, nomencla_num, name, price, socket_type, processors_amount, RAM_type, system_bus_freq):
        InternetStore.__init__(self, nomencla_num, name, price)
        self.socket_type = socket_type
        self.processors_amount = processors_amount
        self.RAM_type = RAM_type
        self.system_bus_freq = system_bus_freq
        
class Processors(InternetStore):
    def __init__(self, nomencla_num, name, price, socket_type, cores_amount, clock_freq, techprocess):
        InternetStore.__init__(self, nomencla_num, name, price)
        self.socket_type = socket_type
        self.cores_amount = cores_amount
        self.clock_freq = clock_freq
        self.techprocess = techprocess
        
class VideoCards(InternetStore):
    def __init__(self, nomencla_num, name, price, vid_chip_clock_freq, fillrate, texture_block_amount):
        InternetStore.__init__(self, nomencla_num, name, price)
        self.vid_chip_clock_freq = vid_chip_clock_freq
        self.fillrate = fillrate
        self.texture_block_amount = texture_block_amount
        
class InternetStoreDB:
    items = dict()
    
    def add_item(self, item):
        self.items[item.nomencla_num] = item
        
    def get_values(self):
        return self.items.values()

    def get_by_number(self, nomencla_num):
        print("Детальна інформація по товару за його номенклатурним номером:")
        for key, value in self.items[nomencla_num].__dict__.items():
            print(key, "-", value)
        
internetStoreDB = InternetStoreDB()
item1 = Motherboards(1, "Asus Prime B460-Plus", 3700, "Socket 1200", 1, "4 x DDR4 DIMM", "Quad Pumping")
item2 = Processors(2, "Intel Core i7-10700K 3.8GHz", 11700, "Socket 1200", 8, "3800 МГц", "14 нм")
item3 = VideoCards(3, "Asus PCI-Ex Radeon RX 6900 XT ROG Strix LC 16GB GDDR6", 63000, 
                   "16000 МГц", "1200 МГц", 20)

internetStoreDB.add_item(item1)
internetStoreDB.add_item(item2)
internetStoreDB.add_item(item3)

print("Перелік комплектуючих інформаційної системи Інтернет-магазину:")
for item in internetStoreDB.get_values():
    a = dict()
    a = item.__dict__
    for key, value in a.items():
        print(key, "-", value)
    print()
    
internetStoreDB.get_by_number(1)
```
