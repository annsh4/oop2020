1. За допомогою конструктору класу, створити клас (будь-якої предметної галузі) з 3-ма атибутами і 3-ма методами, що можуть змінювати їх значення.
2. Створити 3 екземпляри класу.
3. Роздрукувати властивості всіх екзмемплярів.
4. У 1-му екземплярі змінити 1-е значення атирибуту базового класу
5. У 2-му екземплярі змінити 2-е значення атирибуту базового класу
6. У 3-му екземплярі змінити 3-е значення атирибуту базового класу
7. Роздрукувати всі значення атирибутів всіх екземплярів
8. Провести інтерпретацію.

```python
class BloodType:
    def __init__(self, group, rhesus, donor):
        self.group = group
        self.rhesus = rhesus
        self.donor = donor
        
    def change_group(self, group):
        self.group = group
        
    def change_rhesus(self, rhesus):
        self.rhesus = rhesus
        
    def change_donor(self, donor):
        self.donor = donor
        
    def print_info(self):
        print("Група крові:", self.group)
        print("Резус фактор:", self.rhesus)
        print("Донор:", self.donor, "\n")

person1 = BloodType("A", "-", "0-")
person2 = BloodType("B", "+", "0+")
person3 = BloodType("AB", "-", "AB-")

person1.print_info()
person2.print_info()
person3.print_info()        
print("----------------")

person1.change_group("0")
person2.change_rhesus("-")
person3.change_donor("B-")

person1.print_info()
person2.print_info()
person3.print_info()  
```

Індивідуальне завдання №4

1. Cтворити програму, яка визначає наявність товарів на складі, що характризуютьс наступними характристиками (атрибути класу): назва товару, тип товару, код товару, кількість. Зберігання товарів здійснюється у залежності від коду товару: 1 - промисловий склад, 2 - склад харчових продуктів, 3 - склад хімічних речовин, 4 - оптовий склад, 5 - резервний склад
2. Представити постачання 5 різних товарів на склади (створити 5 екземплярів класу) різного типу.
3. Застосувати метод зміни кількості товарів на складі
4. Роздрукувати залишки на складі
5. Провести інтерпретацію.

```python
class Product:
    def __init__(self, name, product_type, code, quantity):
        self.name = name
        self.type = product_type
        self.code = code
        self.quantity = quantity
        
    def __repr__(self):
        return '{} - {} (код: {}, кількість: {})'.format(self.name, self.type, self.code, self.quantity)
    
class Storage:
    def __init__(self, name, code):
        self.name = name
        self.code = code
        self.products = list()
    
    def add_product(self, product):
        self.products.append(product)
        
    def __str__(self):
        return '{} із кодом {} має товари: {}'.format(self.name, self.code, self.products)
        
class StoragesList:
    storages = dict()
    
    def __init__(self, st1, st2, st3, st4, st5):
        self.storages[st1.code] = st1
        self.storages[st2.code] = st2
        self.storages[st3.code] = st3
        self.storages[st4.code] = st4
        self.storages[st5.code] = st5 
        
    def supply_product(self, product):
        product_code = product.code
        if product_code in self.storages.keys():
            self.storages[product_code].add_product(product)

prod1 = Product("Диван", "побутовий товар", 1, 1)
prod2 = Product("Книги", "поліграфія", 4, 100)
prod3 = Product("Рис", "крупа", 2, 50)
prod4 = Product("Ноутбук", "електроніка", 5, 2)
prod5 = Product("Іграшка", "побутовий товар", 5, 8)

st1 = Storage("Промисловий склад", 1)
st2 = Storage("Склад харчових продуктів", 2)
st3 = Storage("Склад хімічних речовин", 3)
st4 = Storage("Оптовий склад", 4)
st5 = Storage("Резервний склад", 5)

storages = StoragesList(st1, st2, st3, st4, st5)

storages.supply_product(prod1)
storages.supply_product(prod2)
storages.supply_product(prod3)
storages.supply_product(prod4)
storages.supply_product(prod5)

print(st1)
print()
print(st2)
print()
print(st3)
print()
print(st4)
print()
print(st5)
```

Індивідуальне завдання №5

Використовуючи композицію

Модифікувати програму надану в лекці:
 - створити клас фарбування підлоги (колір на вибір);
 - створити клас фарбування стелі (білий колір), враховуючи квадратну люстру розміром 1м*1м, 
   яку не потрібно фарбувати;
 - врахувати 2 дверей і 3 вікна;
 - вивести площу фарбування: підлоги, стелі, стін 
Провести інтерпретацію

```python
class Floor:
    def __init__(self, x, y, color):
        self.floor_area = x * y
        self.color = color
    
    def print_info(self):
        print("Площа пофарбованої підлоги:", self.floor_area, "кв.м")
        print("Колір фарби для підлоги:", self.color)
        
class Сeiling:
    def __init__(self, x, y):
        self.ceiling_area = x * y
        self.color = "Білий"
        
    def chandelier(self):
        self.chandelier = Chandelier(1, 1)
    
    def painted_ceiling(self):
        self.painted_ceiling = self.ceiling_area - self.chandelier.area
        
    def print_info(self):
        print("Площа стелі:", self.ceiling_area, "кв.м")
        print("Площа пофарбованої стелі:", self.painted_ceiling, "кв.м")
        print("Колір фарби для стелі:", self.color)
        
class Chandelier:
    def __init__(self, x, y):
        self.area = x * y
        
class Walls:
    def __init__(self, x, y, z, color):
        self.wall = 2 * z * (x + y)
        self.color = color
        
    def windows_doors(self, win_width, win_length, door_width, door_length):
        self.window = Windows_Doors(win_width, win_length)
        self.door = Windows_Doors(door_width, door_length)
        self.windows_number = 3
        self.doors_number = 2
        
    def painted_walls(self):
        self.painted_walls = self.wall - self.window.area * self.windows_number - self.door.area * self.doors_number
            
    def print_info(self):
        print("Площа стін:" , self.wall, "кв.м")
        print("Площа пофарбованих стін:", self.painted_walls, "кв.м")
        print("Колір фарби для стін:", self.color)
            
class Windows_Doors:
    def __init__(self, x, y):
        self.area = x * y

class Room:
    def __init__(self, x, y, z, wall_color, floor_color):
        self.x = x
        self.y = y
        self.z = z
        
        self.wall_color = wall_color
        self.floor_color = floor_color
        
    def print_wall_info(self):
        self.wall = Walls(self.x, self.y, self.z, self.wall_color)
        self.wall.windows_doors(1, 3, 1, 2)
        self.wall.painted_walls()
        self.wall.print_info()
        print()
        
    def print_floor_info(self):
        self.floor = Floor(self.x, self.y, self.floor_color).print_info()
        print()
        
    def print_ceiling_info(self):
        self.ceiling = Сeiling(self.x, self.y)
        self.ceiling.chandelier()
        self.ceiling.painted_ceiling()
        self.ceiling.print_info()
        print()

my_room = Room(1, 1, 15, 'Зелений', 'Сірий')
my_room.print_wall_info()
my_room.print_floor_info()
my_room.print_ceiling_info()
```
